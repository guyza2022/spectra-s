```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('raw_spectra.csv')
```

```python
print(df.head(),df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
#df['Serovar'] = label_encoder.fit_transform(df['Serovar'])
srv = df.Serovar
```

```python
df = df.drop(columns=['Serovar'])
```

```python
len(df.columns)
```

```python
n_del = len(df.columns)%10
for i in range(n_del):
    if i%2 == 0:
        df = df.iloc[:,1:]
    else:
        df = df.iloc[:,:-1]
```

```python
len(df.columns)
print(df.head())
```

```python
bin_size = 10
bin_data = []
for i, row in df.iterrows():
    vals = list(row.values)
    val_chunks = [vals[i:i + bin_size] for i in range(0, len(vals), bin_size)]
    val_mean_chunk = [np.average(x) for x in val_chunks]
    bin_data.append(val_mean_chunk)
cols = list(df.columns.astype(float))
col_chunks = [cols[i:i + bin_size] for i in range(0, len(cols), bin_size)]
col_mean_chunk = [np.round(np.average(x),1) for x in col_chunks]
binned_df = pd.DataFrame(bin_data,columns=col_mean_chunk)
```
